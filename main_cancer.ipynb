## Implementação de uma MLP para realizar a predição se um câncer é maligno ou benigno

## Tratamento do Dataset

```python
import pandas as pd
```

#### Como podemos observar, o dataset tem como entrada:

```python
data = pd.read_csv('datasets/Breast_cancer_wisconsin_dataset.csv')
data
```

```python
##Exclusão do campo Id, pois não será relevante para o algoritmo
data.drop('ID',inplace=True,axis=1)
```

```python
data.isnull().sum()## como podemos observar, não há elementos nulos em nenhuma das campos
```

```python
## Agora será substituido o valor 2 do campo class por 0(benigno) e 4 por 1(maligno)
for i, row in data.iterrows():
    if row['Class']==2:
        data.loc[i,'Class'] = 0
    else:
        data.loc[i,'Class'] = 1
```

```python
import numpy as np
def correcao_valores(df):
    result = df.copy()
    for feature_name in result.columns:
        #print(feature_name)
        mean = np.mean(pd.to_numeric(df[feature_name], errors='coerce'))
        
        for i, row in result.iterrows():
            aux = row[feature_name]
            if not str(row[feature_name]).isdigit():
                result.loc[i,feature_name] =mean
                aux = mean
                
            result.loc[i,feature_name]=float(aux)
    return result
df = correcao_valores(data)
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
dn = normalize(df)
```
